```python
from bs4 import BeautifulSoup
import requests
import pandas as pd

# Define the URL for Wikipedia page
url = 'https://en.wikipedia.org/wiki/List_of_largest_Brazilian_companies'

page = requests.get(url)

# Parse the page content
soup = BeautifulSoup(page.text, 'html')

# Find the table containing the list of largest Brazilian companies
soup.find_all('table')[1]

table = soup.find_all('table')[1]

# Extract table headers
world_titles = table.find_all('th')

world_table_titles = [title.text.strip() for title in world_titles]

# Create an empty DataFrame with headers
df = pd.DataFrame(columns = world_table_titles)

column_data = table.find_all('tr')

# Extract each row in the table
for row in column_data[1:]: # Skip the header row
    row_data = row.find_all('td')
    individual_row_data  = [data.text.strip() for data in row_data]

    length = len(df)
    df.loc[length] = individual_row_data

df
```

```python
# Export DataFrame to CSV
df.to_csv('Largest Brazilian Companies According to Wikipedia.csv', index=False)
print("Data exported successfully to csv")
```
